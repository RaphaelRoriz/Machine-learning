```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#useful functions

def get_day_of_week(day):
    
    #dias da semana para serem buscados por indice 
    DAYS = {"Monday":0,"Tuesday":1,"Wednesday":2,"Thursday":3,"Friday":4,"Saturday":5,"Sunday":6}
    return str(DAYS[day])

def get_day_of_month(date):
    date = date.split()
    aux = date[0].split('-')
    return int(aux[2])
    
    
```

```python

data = pd.read_csv("data.csv") 
data = data.drop(columns=["INCIDENT_NUMBER","OFFENSE_CODE","OFFENSE_CODE_GROUP","OFFENSE_DESCRIPTION",
                         "REPORTING_AREA","SHOOTING","UCR_PART","STREET","Location"])
data = data.drop(0)#dropping the test row

#removing rows with null values.
#data = data[np.isfinite(data['OCCURRED_ON_DATE'])] HOW TO REMOVE NULL DATETIME VALUES?
data = data[np.isfinite(data['YEAR'])]
data = data[np.isfinite(data['MONTH'])]
#data = data[np.isfinite(data['DAY_OF_WEEK'])]HOW TO REMOVE NULL STRING VALUES?
data = data[np.isfinite(data['HOUR'])]
data = data[np.isfinite(data['Lat'])]
data = data[np.isfinite(data['Long'])]

#smaller dataset for tests usage
testedata = data.iloc[500:510].copy()


#print(testedata)

#List to store the indexes of the days of the week
DAY_OF_WEEK_INDEX = []
DAY_OF_MONTH= []

#get the indexes of the days of week and the days of month in the dataset
for row in data.itertuples():
    if row.DISTRICT is None:
        data.drop(row)
    else:
        DAY_OF_WEEK_INDEX.append(int(get_day_of_week(row.DAY_OF_WEEK)))
        DAY_OF_MONTH.append(get_day_of_month(row.OCCURRED_ON_DATE))
    
#cheching if exists missing values in the dataset
print(data.isnull().sum())

#add the days indexes column to the dataset
data['DAY_OF_WEEK_INDEX'] = DAY_OF_WEEK_INDEX
data['DAY_OF_MONTH'] = DAY_OF_MONTH


#drop the column with the name of the days.
data = data.drop(columns="DAY_OF_WEEK")
data = data.drop(columns="OCCURRED_ON_DATE")

data = data[['DAY_OF_MONTH','MONTH','YEAR','DAY_OF_WEEK_INDEX','HOUR','Lat','Long','DISTRICT']]


testedata = data.iloc[250:400].copy()
print(testedata)

"""
#visualizing the data 
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(testedata['DAY_OF_WEEK_INDEX'], bins=30)
plt.show()

#visualizing correlactiin in the data
correlation_matrix = testedata.corr().round(2)
# annot = True to print the values inside the square
sns.heatmap(data=correlation_matrix, annot=True)
"""


```
